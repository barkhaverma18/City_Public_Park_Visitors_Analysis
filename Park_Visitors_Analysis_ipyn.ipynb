```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/sample_data/Park_Visitors.csv")
df
```

```python
df.describe()
```

```python
max_visitors = df['Visitors'].max()
peak_hour = df[df['Visitors']==max_visitors]
print("Maximum Visitors:",max_visitors)
print("Peak Hour:",peak_hour)
```

```python
plt.figure(figsize=(10,5))
plt.bar(df["Time"],df["Visitors"],color="Green")
plt.xticks(rotation=90)
plt.xlabel("Time")
plt.ylabel("Number of Visitors")
plt.title("City Public Park Visitors Analysis")
plt.show()
```
